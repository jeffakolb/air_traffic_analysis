```python
import csv
import pandas as pd
```

```python
traffic_df = pd.read_csv("data/air_segment_data_2017_12.csv",
                 sep=',',                 
                 quotechar='"',
                 header=0,
                )
```

```python
traffic_df.tail()
```

```python
aircraft_df = pd.read_csv('data/aircraft_type.csv')
```

```python
aircraft_df.head()
```

```python
df = pd.merge(traffic_df,aircraft_df,left_on="AIRCRAFT_TYPE",right_on="Code")
fields_to_drop = ['Code','AIRCRAFT_TYPE','YEAR','MONTH','Unnamed: 11']
df = df.drop(fields_to_drop,axis=1)
```

```python
df = df.query('DEPARTURES_SCHEDULED>0')
```

```python
freight_carriers = ['ABX','FX','7S','5X']

df_passenger = df.loc[~df['CARRIER'].isin(freight_carriers)]
```

```python
df.query('CARRIER=="UA"&ORIGIN=="BNA"&DEST=="SFO"')
```

```python
# find top destinations from LAX by passenger count
df_passenger.groupby(['ORIGIN','DEST','CARRIER','CARRIER_NAME']).sum().loc['LAX'] \
.sort_values(['DEPARTURES_SCHEDULED'],ascending=False)
```

```python
df_passenger[df_passenger['Description'].str.contains('767')].sort_values('PASSENGERS',ascending=False)
```
